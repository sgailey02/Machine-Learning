```
#8 bedroom => 100K + 8 * 50K => 500K write a program to predict the price of an 8bdr house
```

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
#y = 100k + number_of_bedroom * 50k
def house_model(y_new):
    xs = np.array([1.0, 2.0, 5.0, 6.0, 8.0, 10.0], dtype=float)
    ys = np.array([1.5, 2.0, 3.5, 4.0, 5.0, 6.0], dtype=float) 
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=200, batch_size=4)  # training
    # steps = length of the dataset / batch_size
    model.save('houseprice_model.h5')
    return model.predict(y_new)[0]
```

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
prediction = house_model([11.0])[0]
print(prediction * 100, "k")
```
